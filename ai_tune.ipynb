## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer,
    DataCollatorForLanguageModeling,
    BitsAndBytesConfig
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
import torch
import numpy as np
from typing import Dict, List
import os

 # Change to your preferred model
chat_file = "/content/Ai/note/cleane.txt"
max_length = 512
output_dir = "./finetuned_model2"
# ==================

# 1️⃣ Load tokenizer + special tokens
print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(
    base_model_id,
    use_fast=False,
    trust_remote_code=True
)

# Add padding token if missing
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.pad_token_id = tokenizer.eos_token_id

# Add special tokens for user/assistant
special_tokens = {
    "additional_special_tokens": ["<|user|>", "<|assistant|>", "<|endoftext|>"]
}
num_added_tokens = tokenizer.add_special_tokens(special_tokens)
print(f"Added {num_added_tokens} special tokens")

# Get special token IDs
USER_TOKEN_ID = tokenizer.convert_tokens_to_ids("<|user|>")
ASSISTANT_TOKEN_ID = tokenizer.convert_tokens_to_ids("<|assistant|>")
EOS_TOKEN_ID = tokenizer.eos_token_id

# 2️⃣ Load .txt dataset into structured conversations
def load_txt_dataset(path):
    """Load conversations from text file with proper role tracking"""
    conversations = []
    current_convo = []

    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    i = 0
    while i < len(lines):
        line = lines[i].strip()

        if line == "<|user|>":
            i += 1
            while i < len(lines) and not lines[i].strip():
                i += 1
            if i < len(lines):
                user_msg = []
                while i < len(lines) and lines[i].strip() not in ["<|user|>", "<|assistant|>"]:
                    if lines[i].strip():
                        user_msg.append(lines[i].strip())
                    i += 1
                if user_msg:
                    current_convo.append({"role": "user", "content": " ".join(user_msg)})
                i -= 1

        elif line == "<|assistant|>":
            i += 1
            while i < len(lines) and not lines[i].strip():
                i += 1
            if i < len(lines):
                assistant_msg = []
                while i < len(lines) and lines[i].strip() not in ["<|user|>", "<|assistant|>"]:
                    if lines[i].strip():
                        assistant_msg.append(lines[i].strip())
                    i += 1
                if assistant_msg:
                    current_convo.append({"role": "assistant", "content": " ".join(assistant_msg)})

                if i < len(lines) and lines[i].strip() == "<|user|>":
                    if len(current_convo) >= 2:
                        conversations.append({"messages": current_convo})
                    current_convo = []
                i -= 1

        i += 1

    if current_convo and len(current_convo) >= 2:
        conversations.append({"messages": current_convo})

    return conversations

# 3️⃣ Format conversation with EOS tokens before role changes
def format_conversation_with_masking(messages):
    """
    Format conversation and create labels mask.
    Returns both the formatted text and positions where assistant responses start/end.
    """
    text = ""
    assistant_ranges = []  # Store (start, end) character positions of assistant responses

    for i, msg in enumerate(messages):
        # Add EOS token before role change (except for first message)
        if i > 0:
            text += tokenizer.eos_token + "\n"

        if msg["role"] == "user":
            text += f"<|user|>\n{msg['content']}\n"
        elif msg["role"] == "assistant":
            start_pos = len(text)  # Mark start of assistant response
            text += f"<|assistant|>\n{msg['content']}\n"
            end_pos = len(text)  # Mark end of assistant response
            assistant_ranges.append((start_pos, end_pos))

    # Add final EOS token
    text += tokenizer.eos_token

    return text, assistant_ranges

# 4️⃣ Create packed batches with proper masking
def create_packed_batches_with_masking(conversations, max_length=512):
    """
    Pack multiple conversations into batches and track assistant response positions.
    """
    batches = []
    batch_masks = []  # Store masking info for each batch
    current_batch = ""
    current_masks = []
    current_tokens = 0

    for convo in conversations:
        convo_text, assistant_ranges = format_conversation_with_masking(convo["messages"])

        # Tokenize to check length
        tokens = tokenizer.encode(convo_text, add_special_tokens=True)
        token_count = len(tokens)

        if token_count > max_length:
            # Truncate single long conversation
            tokens = tokens[:max_length]
            truncated_text = tokenizer.decode(tokens, skip_special_tokens=False)
            batches.append(truncated_text)
            # For truncated text, we'll mask everything except assistant responses
            batch_masks.append("full")  # Special flag for full conversation masking
            current_batch = ""
            current_masks = []
            current_tokens = 0
        elif current_tokens + token_count <= max_length:
            # Add to current batch
            offset = len(current_batch)
            current_batch += convo_text
            current_tokens += token_count
            # Adjust assistant ranges for the offset in the batch
            adjusted_ranges = [(start + offset, end + offset) for start, end in assistant_ranges]
            current_masks.extend(adjusted_ranges)
        else:
            # Save current batch and start new one
            if current_batch:
                batches.append(current_batch)
                batch_masks.append(current_masks)

            current_batch = convo_text
            current_masks = assistant_ranges.copy()
            current_tokens = token_count

    # Add remaining batch
    if current_batch:
        batches.append(current_batch)
        batch_masks.append(current_masks)

    return batches, batch_masks

# 5️⃣ Advanced tokenization with proper label masking
def tokenize_batch_with_masking(batch_text, mask_info):
    """
    Tokenize batch and create labels that only train on assistant responses.
    """
    # Tokenize the batch
    encoding = tokenizer(
        batch_text,
        truncation=True,
        padding="max_length",
        max_length=max_length,
        return_tensors="pt"
    )

    # Create labels
    labels = encoding["input_ids"].clone().squeeze()

    # Mask everything by default
    labels[:] = -100

    if mask_info == "full":
        # For truncated conversations, unmask only assistant tokens
        input_ids = encoding["input_ids"].squeeze()

        # Find assistant response sections
        assistant_mask = torch.zeros_like(input_ids, dtype=torch.bool)
        in_assistant = False

        for i in range(len(input_ids)):
            if input_ids[i] == ASSISTANT_TOKEN_ID:
                in_assistant = True
            elif input_ids[i] == USER_TOKEN_ID:
                in_assistant = False
            elif in_assistant and input_ids[i] != tokenizer.pad_token_id:
                assistant_mask[i] = True

        # Unmask assistant responses
        labels[assistant_mask] = input_ids[assistant_mask]
    else:
        # Use character-level mask info to determine token-level masking
        char_to_token = encoding.char_to_token

        # Create a character-level mask
        char_mask = np.zeros(len(batch_text), dtype=bool)
        for start, end in mask_info:
            char_mask[start:end] = True

        # Convert character mask to token mask
        input_ids = encoding["input_ids"].squeeze()
        for i in range(len(input_ids)):
            if input_ids[i] == tokenizer.pad_token_id:
                continue

            # Get character span for this token
            token_start = encoding.token_to_chars(i)
            if token_start is not None:
                char_idx = token_start.start
                if char_idx < len(char_mask) and char_mask[char_idx]:
                    # This token is part of an assistant response
                    labels[i] = input_ids[i]

    return {
        "input_ids": encoding["input_ids"].squeeze(),
        "attention_mask": encoding["attention_mask"].squeeze(),
        "labels": labels
    }

## 6️⃣ Custom Data Collator for better batching
class CustomDataCollator:
    def __init__(self, tokenizer):
        self.tokenizer = tokenizer

    def __call__(self, features):
        # Convert lists to tensors if needed
        input_ids = []
        attention_mask = []
        labels = []

        for f in features:
            if isinstance(f["input_ids"], list):
                input_ids.append(torch.tensor(f["input_ids"]))
            else:
                input_ids.append(f["input_ids"])

            if isinstance(f["attention_mask"], list):
                attention_mask.append(torch.tensor(f["attention_mask"]))
            else:
                attention_mask.append(f["attention_mask"])

            if isinstance(f["labels"], list):
                labels.append(torch.tensor(f["labels"]))
            else:
                labels.append(f["labels"])

        batch = {
            "input_ids": torch.stack(input_ids),
            "attention_mask": torch.stack(attention_mask),
            "labels": torch.stack(labels)
        }
        return batch
# ===== MAIN DATASET PREPARATION =====
print("Loading conversations from file...")
conversations = load_txt_dataset(chat_file)
print(f"Loaded {len(conversations)} conversations")

# Show sample conversations
if conversations:
    print("\n===== Sample Conversations =====")
    for i, convo in enumerate(conversations[:2]):
        print(f"\nConversation {i+1}:")
        for msg in convo["messages"]:
            role_emoji = "👤" if msg["role"] == "user" else "🤖"
            content_preview = msg['content'][:100] + "..." if len(msg['content']) > 100 else msg['content']
            print(f"  {role_emoji} {msg['role']}: {content_preview}")

print("\n===== Creating Packed Batches with Masking =====")
packed_batches, batch_masks = create_packed_batches_with_masking(conversations, max_length=max_length)
print(f"Created {len(packed_batches)} packed batches")

# Tokenize all batches
print("\n===== Tokenizing Batches =====")
tokenized_examples = []
for i, (batch_text, mask_info) in enumerate(zip(packed_batches, batch_masks)):
    tokenized = tokenize_batch_with_masking(batch_text, mask_info)
    tokenized_examples.append(tokenized)

    if i == 0:
        print(f"\n📊 First Batch Statistics:")
        print(f"{'='*50}")

        # Token statistics
        non_pad_tokens = (tokenized["input_ids"] != tokenizer.pad_token_id).sum()
        trainable_tokens = (tokenized["labels"] != -100).sum()

        print(f"Token Usage:")
        print(f"  • Total tokens: {len(tokenized['input_ids'])}")
        print(f"  • Non-padding tokens: {non_pad_tokens.item()}")
        print(f"  • Trainable tokens (assistant): {trainable_tokens.item()}")
        print(f"  • Training efficiency: {(trainable_tokens.item() / non_pad_tokens.item() * 100):.1f}%")
        print(f"  • Padding: {(1 - non_pad_tokens.item() / len(tokenized['input_ids'])) * 100:.1f}%")

# Create HuggingFace Dataset
dataset_dict = {
    "input_ids": [ex["input_ids"] for ex in tokenized_examples],
    "attention_mask": [ex["attention_mask"] for ex in tokenized_examples],
    "labels": [ex["labels"] for ex in tokenized_examples]
}
train_dataset = Dataset.from_dict(dataset_dict)

print(f"\n✅ Final Dataset Statistics:")
print(f"  • Total training examples: {len(train_dataset)}")
print(f"  • Tokens per example: {max_length}")
print(f"  • Total conversations processed: {len(conversations)}")

# Save dataset
train_dataset.save_to_disk("./fine_tuning_dataset")
print(f"\n💾 Dataset saved to ./fine_tuning_dataset")



````
